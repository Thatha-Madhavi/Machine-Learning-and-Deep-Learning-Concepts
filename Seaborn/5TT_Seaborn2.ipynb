```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

sns.barplot(x='class', y='fare', hue= 'sex', ci=None, data = titanic)

plt.xlabel('class')
plt.ylabel("Fare")
plt.title("Fare by class and sex")
plt.show()
```
