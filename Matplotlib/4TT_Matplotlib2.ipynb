```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.plot(x,y)
plt.title("Sports Data")
plt.xlabel("Average pulse")
plt.ylabel("Calorie Burnage")
plt.grid(axis ='y')
plt.show()
```

```python
#plt1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

#plt2
x = np.array([0,1,2,3])
y = np.array([10,20,30,40])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.show()
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])

plt.scatter(x,y, color='blue')

x1 = np.array([80,85,90,95,100,125,110,115,120,145])
y1 = np.array([240,250,260,277,280,290,300,310,320,330])

plt.scatter(x1,y1, color='red')
plt.show()
```

```python
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300])

plt.scatter(x,y, s=sizes)
plt.show()
```

```python
x = np.array(['Madhu', 'B', 'C', 'D'])
y = np.array([3,8,1,10])

plt.bar(x,y)
plt.show()
```

```python
x = np.array(['Madhu', 'B', 'C', 'D'])
y = np.array([3,8,1,10])

plt.barh(x,y, color ='green', height=0.1)
plt.show()
```

```python
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

Pie charts in Matplotlib

```python
y = np.array([35,25,25,15])
mylabels = ['Apples', 'Bananas', 'C', 'D']
myexplode = [0.2,0.1,0,0]
mycolors = ["black", 'r', 'blue', 'green' ]
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True, colors = mycolors)
plt.legend(title = "Four Fruits")
plt.show()
```
