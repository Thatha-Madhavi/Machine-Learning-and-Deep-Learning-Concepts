```python
xpoints = np.array([0,1,2,3])
ypoints = np.array([3,8,9,20])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([3,8,9,20])
plt.plot(ypoints, marker= "*")
plt.show()
```

```python
ypoints = np.array([3,8,9,20])
plt.plot(ypoints, "o:r")
plt.show()
```

```python
plt.plot(ypoints, marker='o', ms=10, mec= 'r', mfc='green')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, ls = ':')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, ls = '--')
plt.show()
```

```python
ypoints = np.array([3,8,1,10])
plt.plot(ypoints, color = 'r')
plt.show()
```
