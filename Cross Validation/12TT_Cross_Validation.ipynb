```python
import numpy as np
import pandas as pd
from scipy.special import comb
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import (
    KFold,
    RepeatedKFold,
    LeaveOneOut,
    LeavePOut,
    StratifiedKFold,
    cross_val_score,
    cross_validate,
    train_test_split
)
```

```python
X,y = load_breast_cancer(return_X_y=True, as_frame = True)
```

```python
y.value_counts() / len(y)
```

```python
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)
X_train.shape, X_test.shape
```

```python
# K-Fold Cross-Validation
logit = LogisticRegression(penalty = 'l2', C=10, solver = 'liblinear', random_state = 4, max_iter = 10000)
kf = KFold(n_splits = 5, shuffle = True, random_state = 4)

clf = cross_validate(
    logit,
    X_train,
    y_train,
    cv = kf,
    scoring = 'accuracy',
    return_train_score=True,
)
```

```python
clf['test_score']
```

```python

clf['train_score']
```

```python
print('mean train set accuracy:', np.mean(clf['train_score']), '+-', np.std(clf['train_score']))
```

```python
print('mean test set accuracy:', np.mean(clf['test_score']), '+-', np.std(clf['test_score']))
```

```python
#Repeated k-Fold Cross-Validation
logit = LogisticRegression(
    penalty = 'l2', C= 1, solver = 'liblinear', random_state = 4, max_iter=10000
)
rfc = RepeatedKFold(n_splits = 5, shuffle = True , random_state = 4)

clf = cross_validate(
    logit,
    X_train,
    y_train,
    cv = kf,
    scoring = 'accuracy',
    return_train_score=True,
)
```

```python
print('mean train set accuracy:', np.mean(clf['train_score']), '+-', np.std(clf['train_score']))
```

```python
print('mean test set accuracy:', np.mean(clf['test_score']), '+-', np.std(clf['test_score']))
```

```python
# Leave-One-Out CV
logit = LogisticRegression(
    penalty = 'l2', C= 1, solver = 'liblinear', random_state = 4, max_iter=10000
)
Loo = LeaveOneOut()
print('We expect as many metrics as date in the train set:', len(X_train))
clf = cross_validate(
    logit,
    X_train,
    y_train,
    cv = Loo,
    scoring = 'accuracy',
    return_train_score=True,
)
```

```python
# Leave-P-Out CV
logit = LogisticRegression(
    penalty = 'l2', C= 1, solver = 'liblinear', random_state = 4, max_iter=10000
)
Lpo = LeavePOut(p=2)
X_train_small = X_train.head(100)
y_train_small = y_train.head(100)
print('We Expect:', comb(100,2), 'metrics')
clf = cross_validate(
    logit,
    X_train,
    y_train,
    cv = Lpo,
    scoring = 'accuracy',
    return_train_score=True,
)
```

```python
print('mean train set accuracy:', np.mean(clf['train_score']), '+-', np.std(clf['train_score']))
```

```python
print('mean test set accuracy:', np.mean(clf['test_score']), '+-', np.std(clf['test_score']))
```

```python
#Stratified K-Fold CV
logit = LogisticRegression(
    penalty = 'l2', C= 1, solver = 'liblinear', random_state = 4, max_iter=10000
)
rfc = StratifiedKFold(n_splits = 5, shuffle = True , random_state = 4)

clf = cross_validate(
    logit,
    X_train,
    y_train,
    cv = kf,
    scoring = 'accuracy',
    return_train_score=True,
)
```
