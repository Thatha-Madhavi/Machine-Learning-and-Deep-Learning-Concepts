```python
#Standard Deviation
import numpy
speed=[32,111,135,28,59,77,86]
x= numpy.mean(speed)
y = numpy.std(speed)
print(x)
print(y)
```

```python
#Varience
speed = [32,111,138,28,59,77,86]
x= numpy.var(speed)
print(x)
```

```python
#Percentile
ages = [5,31,43,48,50,41,7,11,15,39,80,82,32,2,8,6,25,36,27,61,31]
x = numpy.percentile(ages,90)
print(x)
```

```python
#Regression
from scipy import stats
x=[5,7,8,7,2,17,2,9,4,11,12,9,6]
y=[99,86,87,88,111,86,103,87,94,78,77,85,86]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x,y)
```

```python
def myfun(x):
  return slope * x + intercept
```

```python
mymodel = list(map(myfun,x))
```

```python
plt.scatter(x,y)
plt.plot(x,mymodel)
plt.show()
```
